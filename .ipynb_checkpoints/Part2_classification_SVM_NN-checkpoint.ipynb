```python
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_learning_curves
from sklearn.metrics import classification_report, confusion_matrix
from keras.models import Sequential 
from keras.layers import Dense,Activation,Dropout 
from keras.layers.normalization import BatchNormalization 
from keras.utils import np_utils
from keras.optimizers import Adam
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import normalize
```

```python
df =  pd.read_csv("dataForPart2.csv")
df.head()
```

```python
df=df.drop(['Unnamed: 0'], axis=1)
```

SVM:
    -------------

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
sc = MinMaxScaler(feature_range = (0, 1))
X = sc.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
model=SVC()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
f, ax = plt.subplots(figsize=(10, 9))
sns.heatmap(confusion_matrix(y_test,y_pred), vmax=.8, square=True);
```

```python
print(classification_report(y_test, y_pred))
```

```python
plot_learning_curves(X_train, y_train, X_test, y_test, model)

plt.show()
```

NN:
    -------

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
y = y.reshape(-1, 1)
```

```python
encoder = OneHotEncoder(sparse=False)
y = encoder.fit_transform(y)
```

```python
X=normalize(X,axis=0)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python

model = Sequential()

model.add(Dense(10, input_shape=(4,), activation='relu', name='fc1'))
model.add(Dense(10, activation='relu', name='fc2'))
model.add(Dense(3, activation='softmax', name='output'))


optimizer = Adam(lr=0.001)
model.compile(optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, verbose=2, batch_size=5, epochs=200)
```

```python
results = model.evaluate(X_test, y_test)
```

```python

print('test set loss: {:4f}'.format(results[0]))
print('test set accuracy: {:4f}'.format(results[1]))
```
